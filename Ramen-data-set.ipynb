```python
import os
import pandas as pd
import sqlite3

#note, full source is not necessary if located in same folder
df = pd.read_csv('ramen-ratings-dataset.csv', usecols = ['Brand', 'Variety', 'Style', 'Stars']) 

#Creates a database
db = sqlite3.connect(':memory:')
cursor = db.cursor()

def run_query(query):
    return pd.read_sql_query(query, db)

query = """
    SELECT * 
    FROM ramen 
    WHERE Stars = 5 
    ORDER BY brand ASC 
    LIMIT 50
"""

#Creates new database table
cursor.execute("CREATE TABLE ramen (Brand TEXT, Variety TEXT, Style TEXT, Stars FLOAT);")

#inserts each row of the df into db
for index, row in df.iterrows():
    cursor.execute("INSERT INTO ramen VALUES (?, ?, ?, ?);", 
                  (row['Brand'], row['Variety'], row['Style'], row['Stars']))
#commit changes to db
db.commit()

run_query(query)
    
```
