```python
import os
import pandas as pd
import sqlite3

df = pd.read_csv('ramen-ratings-dataset.csv', usecols = ['Brand', 'Variety', 'Style', 'Stars']) #note, full source is not necessary if located in same folder

#Creates a database
connection = sqlite3.connect(':memory:')
cursor = connection.cursor()

#Creates new database table
cursor.execute("CREATE TABLE ramen (Brand TEXT, Variety TEXT, Style TEXT, Stars FLOAT)")

#inserts each row of the df into db
for index, row in df.iterrows():
    cursor.execute("INSERT INTO ramen VALUES (?, ?, ?, ?);", (row['Brand'], row['Variety'], row['Style'], row['Stars']))

#commit changes to db
connection.commit()

for row in cursor.execute("SELECT * FROM ramen WHERE Stars >= 4;"):
    print(row)
```
