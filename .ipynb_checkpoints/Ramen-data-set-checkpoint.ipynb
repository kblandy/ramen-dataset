```python
import os
import pandas as pd
import sqlite3

#Creates a database (in this case storing into local memory)
db = sqlite3.connect('RamenDemo.db')

#note, full source is not necessary if located in same folder
df1 = pd.read_csv('ramen-ratings-dataset.csv') 

# convert the pandas dataframe to a sqlite table
df1.to_sql("ramen", db, if_exists="replace")

#returns query in table format
def run_query(query):
    return pd.read_sql_query(query, db)

query = """
    SELECT * 
    FROM ramen 
    WHERE variety LIKE "%spicy%"
    AND stars = 5
    ORDER BY brand ASC 
    LIMIT 50
"""

run_query(query)

    
```
